# 5. Impute null values with Mode in Gender Column

```python
df1 = df.isnull()["Gender"].sum()
```

```python
df1
```

```python
# Calculate the mode of the 'Gender' column
df2 = df['Gender'].mode()

# Impute null values with the mode
df['Gender'].fillna(df2,inplace=True)
```

```python
df2
```

# 6. Impute null values with Mode in Self_Employed Column

```python
df3 = df.isnull()["Self_Employed"].sum()
```

```python
df3
```

```python
# Calculate the mode of the 'self_employed' column
self_emp = df['Self_Employed'].mode()

# Impute null values with the mode
df['Self_Employed'].fillna(self_emp,inplace=True)
```

# 7. Impute null values with Median in LoanAmount Column

```python
df.isnull()["LoanAmount"].sum()
```

```python
# Calculate the median of the "LoanAmount" column
median_value = df['LoanAmount'].median()

# Impute null values with the median
df['LoanAmount'].fillna(median_value, inplace=True)
```
