# Q10. How many sharks participated in this show and What were their names? 

```python
# Create an empty list to store the shark names
sharks = []

# Function to add a shark to the list
def add_shark(name):
    sharks.append(name)

# Function to get the count of sharks
def get_shark_count():
    return len(sharks)

add_shark("ashneer_present")
add_shark("anupam_present")
add_shark("aman_present")
add_shark("namita_present")
add_shark("vineeta_present")
add_shark("peyush_present")
add_shark("ghazal_present")
# Get the count of sharks
count = get_shark_count()
print("Number of sharks participated:", count)

# Print the names of all the sharks
print("Shark names are:")
for shark in sharks:
    print(shark)
```
